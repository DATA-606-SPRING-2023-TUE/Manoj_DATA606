The variables "avg_price_per_room" and "leadcolumns" have highly distinctive values, making it difficult to perform data analysis. Sorting these values into clearly defined categories can make visualization easier and more effective.

```python
price_slabs = []
for i in range(len(df)):
    if df.iloc[i,16] == 0:
        price_slabs.append(0)
    elif df.iloc[i,16]>0 and df.iloc[i,16] <= 25:
        price_slabs.append(25)
    elif df.iloc[i,16]>25 and df.iloc[i,16] <= 50:
        price_slabs.append(50)
    elif df.iloc[i,16]>50 and df.iloc[i,16] <= 75:
        price_slabs.append(75)
    elif df.iloc[i,16]>75 and df.iloc[i,16] <= 100:
        price_slabs.append(100)
    elif df.iloc[i,16]>100 and df.iloc[i,16] <= 125:
        price_slabs.append(125)
    elif df.iloc[i,16]>125 and df.iloc[i,16] <= 150:
        price_slabs.append(150)
    elif df.iloc[i,16]>150 and df.iloc[i,16] <= 175:
        price_slabs.append(175)    
    elif df.iloc[i,16]>175 and df.iloc[i,16] <= 200:
        price_slabs.append(200)
    elif df.iloc[i,16]>200 and df.iloc[i,16] <= 225:
        price_slabs.append(225)
    elif df.iloc[i,16]>225 and df.iloc[i,16] <= 250:
        price_slabs.append(250)
    elif df.iloc[i,16]>250 and df.iloc[i,16] <= 275:
        price_slabs.append(275)
    elif df.iloc[i,16]>275 and df.iloc[i,16] <= 300:
        price_slabs.append(300)
    elif df.iloc[i,16]>300 and df.iloc[i,16] <= 325:
        price_slabs.append(325)
    elif df.iloc[i,16]>325 and df.iloc[i,16] <= 350:
        price_slabs.append(350)
    elif df.iloc[i,16]>350 and df.iloc[i,16] <= 375:
        price_slabs.append(375)
    elif df.iloc[i,16]>375 and df.iloc[i,16] <= 400:
        price_slabs.append(400)
    elif df.iloc[i,16]>400 and df.iloc[i,16] <= 425:
        price_slabs.append(425)
    elif df.iloc[i,16]>425 and df.iloc[i,16] <= 450:
        price_slabs.append(450)
    elif df.iloc[i,16]>450 and df.iloc[i,16] <= 475:
        price_slabs.append(475)
    elif df.iloc[i,16]>475 and df.iloc[i,16] <= 500:
        price_slabs.append(500)
    elif df.iloc[i,16]>500 and df.iloc[i,16] <= 525:
        price_slabs.append(525)
    elif df.iloc[i,16]>525 and df.iloc[i,16] <= 550:
        price_slabs.append(550)
```

To make the representation easier, I created pricing slabs for every $25.

```python
df['price_slabs'] = price_slabs
```

I have lead time slabs for every 20 days

```python
lead_time_slabs = []
error = []

for i in range(len(df)):
    lead_time = df.iloc[i,8]
    
    if lead_time == 0:
        lead_time_slabs.append(0)
    elif lead_time > 460:
        error.append(lead_time)
    else:
        lead_time_slabs.append((lead_time // 20 + 1) * 20)
```

```python
df['lead_time_slabs'] = lead_time_slabs
```

```python
df['total_nights'] = df['no_of_weekend_nights'] + df['no_of_week_nights']
```

```python
# Creating a list to encode the booking status as 0 or 1
booking_status_encoded = [0 if status == 'Not_Canceled' else 1 for status in df['booking_status']]

# Create a new dataframe with the encoded booking status column
df_encoded = df.assign(booking_status_encoded=booking_status_encoded)

# Create a heatmap to visualize the correlations between all pairs of columns in the encoded dataframe
plt.figure(figsize=(15,10))
sns.heatmap(df_encoded.corr(), annot = True)
plt.show()
```

We can observe that there is a comparatively strong correlation between the booking status and the lead time. In a classification model, this may be crucial.

```python
plt.figure(figsize=(10, 6))

# Plot the total number of reservations for each lead time slab
df['lead_time_slabs'].value_counts().sort_index().plot.bar()

# Plot the number of canceled reservations for each lead time slab
df[df['booking_status'] == 'Canceled'].lead_time_slabs.value_counts().sort_index().plot.bar(color=['darksalmon'])

# Set the title and legend
plt.title("Cancellations vs Lead Time")
plt.legend(["Total Reservations", "Canceled Reservations"])

# Set the axis labels
plt.xlabel("Lead Time")
plt.ylabel("Number of Reservations")
plt.show()
```

We can observe that as the lead time rises, the concentration of cancellations increases.


```python
# filter the data to only include reservations made in 2018
df_2018 = df[df['arrival_year'] == 2018]

# plot the total number of reservations per month
plt.figure(figsize=(10,6))
df_2018.arrival_month.value_counts(sort=False).plot.line()

# plot the number of cancellations per month
df_cancelled = df_2018[df_2018['booking_status']=='Canceled']
df_cancelled.arrival_month.value_counts().sort_index().plot.line(color=['darksalmon'])

# set the x-axis ticks to be the same as the total number of reservations plot
plt.xticks(df_2018['arrival_month'].value_counts(sort=False).index)

# add title and axis labels
plt.title("Reservations by Month - 2018")
plt.xlabel("Month of the year")
plt.ylabel("Number of reservations")

# add legend
plt.legend(["Total Reservations", "Cancellations"])

plt.show()
```

Summertime has more reservations than wintertime, which is obvious. October is the busiest month for bookings at this hotel.


```python
# filter the data to only include non-cancelled reservations
df_not_cancelled = df[df['booking_status'] != 'Canceled']

# plot the total number of reservations per day
plt.figure(figsize=(10, 6))
df_not_cancelled['arrival_date'].value_counts(sort=False).sort_index().plot.line()

# plot the number of cancellations per day
df_cancelled = df[df['booking_status'] == 'Canceled']
df_cancelled.arrival_date.value_counts().sort_index().plot.line(color=['darksalmon'])

# set the x-axis ticks to be the same as the cancelled reservations plot
plt.xticks(df_cancelled.arrival_date.value_counts().sort_index().index)

# add title and axis labels
plt.title("Reservations and Cancellations by Day")
plt.xlabel("Day of the month")
plt.ylabel("Number of reservations")

# add legend
plt.legend(["Total Reservations", "Cancellations"])

plt.show()
```
